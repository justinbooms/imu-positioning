```python
from __future__ import print_function
from __future__ import division
import math

import pickle
from datetime import datetime, timedelta
import numpy as np

"""Continuous Reading of Sensors

Make continuous readings from the sensors and begin
a take measurements function.

We Believe the Following:
Magnet x: 
Magnet y:
Magnet z:
Euler Heading: Dir w/ 0 being North 90 East, 180 South, 270 West
Euler Roll:
Euler Pitch: Angle up 
Accel x:
Accel y:
Accel z:
Euler x:
Euler y:
Euler z:
Thermometer: Temperature in Celcius
Left Encoder: Odometer of left wheel
Right Encoder: Odometer of right wheel
"""

print("Example program for reading a Dexter Industries Distance Sensor on an I2C port.")


def turnDist(prev_left_encoder, new_left_encoder, prev_right_encoder, new_right_encoder):
    #wheelbase in mm
    wheelbase = 100
    #Test to see if turning left or right
    L_enc_delta = new_left_encoder - prev_left_encoder
    R_enc_delta = new_right_encoder - prev_right_encoder
    
    
    
    
    if L_enc_delta > 15 and R_enc_delta < 5:
        print('turn right')
        #turning right
        turn_angle = L_enc_delta / wheelbase
        #calculate Vector of hypotnuse from one wheel to other
        y_comp = math.sin(turn_angle)*wheelbase
        x_comp = math.cos(turn_angle)*wheelbase
        hyp_vec = [x_comp, y_comp]
        
        #assumes axis made by axle in original location is the x-axis
        
        x_axis_vec = [wheelbase, 0]
        
        #Projects hypotnuse vector onto x-axis
        
        projection_vec = np.dot(hyp_vec, x_axis_vec) / wheelbase
        
        #Calculate distance from outside wheels original location to new location
        
        x_dist_comp = projection_vec - x_comp
        
        outside_turn_distance = (y_comp**2 + x_dist_comp**2)**(1/2)
        
        center_distance_travel = outside_turn_distance/2
        #estimated heading angle delta
        
        cos_angle = x_dist_comp/outside_turn_distance
        
        est_angle = math.acos(cos_angle)
        x = (center_distance_travel, est_angle)
        
        return x
        
    elif R_enc_delta > 15 and L_enc_delta < 5:
        print('turn left')
         #turning left
        turn_angle = R_enc_delta / wheelbase
        #calculate Vector of hypotnuse from one wheel to other
        y_comp = math.sin(turn_angle)*wheelbase
        x_comp = math.cos(turn_angle)*wheelbase
        hyp_vec = [x_comp, y_comp]
        
        #assumes axis made by axle in original location is the x-axis
        
        x_axis_vec = [wheelbase, 0]
        
        #Projects hypotnuse vector onto x-axis
        
        projection_vec = np.dot(hyp_vec, x_axis_vec) / wheelbase
        #Calculate distance from outside wheels original location to new location
        
        x_dist_comp = projection_vec - x_comp
        
        outside_turn_distance = (y_comp**2 + x_dist_comp**2)**(1/2)
        
        center_distance_travel = outside_turn_distance/2
        
        #estimated heading angle delta
        
        cos_angle = x_dist_comp/outside_turn_distance
        
        est_angle = math.acos(cos_angle)
        
        x = (center_distance_travel, est_angle)
        
        return x
    else:
        return
        

        

   
```
